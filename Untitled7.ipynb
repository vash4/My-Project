```python
import pandas as pd
df=pd.read_csv("/content/Fraud.csv")
```

```python
df.head()
```

```python
# dropping feature
df=df.drop(['nameOrig','nameDest','isFlaggedFraud'], axis = 1)
df
```

```python
# storing the fraudulent data into dataframe
fra = df[df['isFraud']==1]
fra
```

```python
# storing the non-fraudulent data into dataframe
non_fra = df[df['isFraud']==0]
non_fra
```

```python
# joining both dataset together
df =pd.concat([fra,non_fra],axis=0)
df
```

```python
#Package Imports
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
#Converting the type column to categorical
df['type'] = df['type'].astype('category')
df
```

```python
#Integer Encoding the 'type' column
encode = LabelEncoder()
encode
```

```python
#Integer encoding the 'type' column
df['type'] = encode.fit_transform(df.type)
df
```

```python
#One hot encoding 
onehot = OneHotEncoder()
onehotencode = onehot.fit_transform(df.type.values.reshape(-1,1)).toarray()
onehot
```

```python
#Adding the one hot encoded variables to the dataset
var= pd.DataFrame(onehotencode, columns = ["type_"+str(int(i)) for i in range(onehotencode.shape[1])])
df = pd.concat([df, var], axis=1)
df
```

```python
#Dropping the original type variable
df = df.drop('type', axis = 1)
df.head()
```

```python
#Checking every column for missing value
df.isnull().any()
```

```python
#Imputing the missing values with a 0
df = df.fillna(0)
df
```

```python
df.to_csv('fraud_prediction.csv')
df
```
